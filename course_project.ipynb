```python
#Let's drop some unnecessary features
activity = activity.drop(["TrackerDistance","LoggedActivitiesDistance","VeryActiveDistance",
                          "ModeratelyActiveDistance","LightActiveDistance","SedentaryActiveDistance"], axis = 1)
activity['ActiveMinutes'] = activity.VeryActiveMinutes+activity.FairlyActiveMinutes
activity
```

```python
#Let's change date data to datetime format to make data processing easier

activity["ActivityDate"] = pd.to_datetime(activity['ActivityDate']).dt.strftime('%m/%d/%Y')
sleep["SleepDay"] = pd.to_datetime(sleep['SleepDay']).dt.strftime('%m/%d/%Y')
weight["Date"] = pd.to_datetime(weight['Date']).dt.strftime('%m/%d/%Y')
activity.info()
```

```python
#Let's drop some unnecessary features

weight = weight.drop(["WeightPounds","Fat","IsManualReport","LogId"], axis = 1)
weight
```

```python
#Let's name all change date info containing columns' names into "Date" to make dataframe joining easier
activity = activity.rename(columns = {"ActivityDate":"Date"})
sleep = sleep.rename(columns = {"SleepDay":"Date"})
```

```python
#First, let's take a closer look at the average stepcount of participants, because it's FitBit's most accurate measurement

df = activity[["Id","TotalSteps"]]

#Let's only take into an account the non-zero values
df = df[df.TotalSteps > 0]
df = df.groupby(['Id']).mean()
print(df.mean().TotalSteps)

#Create a graph
plt.figure(figsize=(16,8))
plt.suptitle("Average stepcount per user id",fontsize=20)
g1 = sns.barplot(data=df,x=df.index,y="TotalSteps",alpha=0.5)
g1.set_xticklabels(df.index, rotation=60, fontsize = 8)
g1.axhline(df.mean().TotalSteps,color="black")
g1.axhline(7000,color="green")
```

```python
#Let's then investigate average sleeping times of the participants

df2 = sleep[["Id","TotalMinutesAsleep"]]

plt.figure(figsize=(16,8))
plt.suptitle("Average sleep time",fontsize=20)
sns.histplot(x = df2.groupby("Id").TotalMinutesAsleep.mean()/60, bins = 12, color="darkblue",kde=True)
print(df2.mean().TotalMinutesAsleep/60)
```

```python
activity['ActiveHours']=activity['ActiveMinutes']/60
activity
```

```python
#Let's join sleeop and activity dataframes to investigate the relationship between sleep and activity

data_joined = activity.set_index(["Id","Date"]).join(sleep.set_index(["Id","Date"]),how="right")
data_joined = data_joined.reset_index()
data_joined["TotalHoursAsleep"] = data_joined.TotalMinutesAsleep/60
data_joined
```

```python
#Create a plot

data_joined['ActivityType'] = np.where(data_joined.ActiveHours > 1, "Active", "Inactive")
plt.figure(figsize=(16,8))
plt.suptitle("Sleep time per activity type",fontsize=20)
sns.histplot(data=data_joined, x="TotalHoursAsleep", hue="ActivityType", bins=12,kde=True,multiple="stack",palette="Blues")
```

```python
plt.figure(figsize=(16,8))
plt.suptitle("Sedentary minutes and total hours asleep",fontsize=20)
sns.scatterplot(x = data_joined.TotalHoursAsleep, y = data_joined.SedentaryMinutes, color="darkblue")
#Here we notice that TotalHoursAsleep and SedentaryMinutes correlate negatively
```

```python
#Finally, let's see how BMI affects activity
data_joined2 = activity.set_index(["Id","Date"]).join(weight.set_index(["Id","Date"]),how="right")
data_joined2 = data_joined2.reset_index()
data_joined2['WeightLabel'] = data_joined2['BMI'].apply(lambda x: 'Overweight' if x >= 25 else ('Normal' if (x < 25 and x >= 18.5) else 'Underweight'))
data_joined2
```

```python
#Let's only take into an account the non-zero values
data_joined2 = data_joined2[data_joined2.TotalSteps > 0]
data_joined2 = data_joined2.groupby(['Id']).mean()
data_joined2['WeightLabel'] = data_joined2['BMI'].apply(lambda x: 'Overweight' if x >= 25 else ('Normal' if (x < 25 and x >= 18.5) else 'Underweight'))
```

```python
data_joined2 = data_joined2.groupby(['WeightLabel']).mean()
data_joined2
```

```python
plt.figure(figsize=(16,8))
plt.suptitle("Average stepcount per weight label",fontsize=20)
sns.barplot(data=data_joined2, x=data_joined2.index, y="TotalSteps", palette="Blues")
```
